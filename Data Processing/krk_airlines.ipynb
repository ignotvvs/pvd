# Flights in Krakow Airport

```python
import pandas as pd
```

```python
airlines = pd.read_csv("krk-airlines.csv")
flights = pd.read_csv("krk-flights.csv")
passengers = pd.read_csv("krk-passengers.csv")
airlines
```

```python
passengers
```

```python
flights
```

```python
flights.loc[:,["flight", "to"]]
```

```python
flights_airlines = pd.merge(flights, airlines, on="airlineid")
flights_airlines.loc[:,["flight", "airline","airplane"]]
```

```python
flights_passengers = pd.merge(flights, passengers, on="flight")
flights_passengers.query("to == 'London'").loc[:, ["name","surname"]].sort_values(by="surname")
```

```python
flights_passengers.query("to == 'Paris' and gender == 'Female'").loc[:, ["name","surname"]].sort_values(by="surname")
```

```python
flights_passengers.query("to == 'Berlin'").groupby("to")["surname"].count()
```

```python
flights_passengers.query("gender == 'Male'").groupby("gender")["surname"].count()
```

```python
passengers.groupby("flight")["surname"].count()
```
